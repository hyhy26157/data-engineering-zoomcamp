```python
import pandas as pd
import datetime
import gzip
```

```python
#dataframe for the trip data
gdf = pd.read_csv('green_tripdata_2019-01.csv.gz', compression='gzip',
                   error_bad_lines=False)

#dataframe for the zone data
df_zones = pd.read_csv('taxi+_zone_lookup.csv')
```

```python
#q3 answers
gdf[(gdf['lpep_pickup_datetime'].str.contains("2019-01-15"))]
```

```python
#fill na for zones data
df_zones = df_zones.fillna('')
```

```python
#q4 create a new column that extract the days from the lpep_pickup_datetime column
gdf['day_pickup'] = gdf['lpep_pickup_datetime'].str.extract("(\d{4}-\d{2}-)(\d{2})")[1]
```

```python
#q4 answer
gdf.groupby(by='day_pickup')['trip_distance'].agg('max').sort_values(ascending=False)
```

```python
#q5 answer
gdf.groupby(by='day_pickup')['passenger_count'].value_counts()
```

```python
#q6 look for zone id for Astoria
df_zones[df_zones['Zone'].str.contains('Astoria')]
```

```python
#q6 agg result to find largest tip amount based on drop location id.
gdf[gdf['PULocationID'] == 7].groupby(by='DOLocationID')['tip_amount'].agg('max').sort_values(ascending=False)
```
